```python
import hashlib
import itertools

HASH_OBJETIVO = "ed770662c27576ae6e22eab82c4798589c6695f5cec25e8ecd4ca9da837fa250"

def encontrar_secuencia(hash_objetivo):
    for perm in itertools.permutations("0123456789"):
        secuencia = "".join(perm)
        hash_calculado = hashlib.sha256(secuencia.encode()).hexdigest()

        if hash_calculado == hash_objetivo:
            return secuencia

    return None

resultado = encontrar_secuencia(HASH_OBJETIVO)

print("Resultado:", resultado)
```
