```python
import hashlib
import itertools

# HASH ROOT
HASH_OBJETIVO = "89b0ea0608ff9a86c5b4e39439921a6a7494e123c194984ab438e7960adb3dd3"

#  Transacciones
transacciones = ["hola", "Hola", "chao", "Chao"]

# ---------- FUNCIONES ----------
def sha256(x):
    return hashlib.sha256(x.encode()).hexdigest()

def merkle_root(tx):
    hashes = [sha256(t) for t in tx]

    while len(hashes) > 1:
        nuevo = []
        for i in range(0, len(hashes), 2):
            if i+1 < len(hashes):
                combinado = hashes[i] + hashes[i+1]
            else:
                combinado = hashes[i] + hashes[i]

            nuevo.append(sha256(combinado))

        hashes = nuevo

    return hashes[0]

# ---------- FUERZA BRUTA ----------
for perm in itertools.permutations(transacciones):
    if merkle_root(list(perm)) == HASH_OBJETIVO:
        print("Orden correcto:", perm)
        break
else:
    print("No se encontró")
```
